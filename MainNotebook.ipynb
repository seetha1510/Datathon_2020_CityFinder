```python
df = pd.read_csv('uscities.csv')
cities1 = df['city']
cities = (df['city']).to_numpy()
population = (df['population']).to_numpy()
density = (df['density']).to_numpy()
state_name = (df['state_name']).to_numpy()
county_name = (df['county_name']).to_numpy()
a = np.matrix([cities, population, density, state_name, county_name])

cities1
```

```python
b = np.vstack(cities)
c = np.vstack(population)
d = np.vstack(density)
e = np.vstack(state_name)
f = np.vstack(county_name)
vmatrix = np.concatenate((b,c,d,e,f),axis=1)
print (vmatrix)
```

```python
response = []

for i in vmatrix:
    if (float(i[1]) <= 49999):
        response = np.append(response, 0)
    elif (float(i[1]) > 50000 and float(i[1]) <= 349999):
        response = np.append(response, 1)
    else:
        response = np.append(response, 2) 
        
response = np.vstack(response)

area_def = []
for i in vmatrix:
    if(float(i[1]) >= 50000 or float(i[2]) >= 1000):
        area_def = np.append(area_def, 2) # urban areas
    elif (float(i[1]) >= 2500 and float(i[1]) < 50000):
    #else:
        area_def = np.append(area_def, 1) # urban clusters
    elif (float(i[1]) < 2500 or float(i[2]) < 500):
        area_def = np.append(area_def, 0) # rural areas
area_def = np.vstack(area_def)
        
result = np.concatenate((vmatrix, response,area_def), axis=1)
```
