# Creating master dataset with price, diversity and other data

```python
#importing needed libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
uscities = pd.read_csv('uscities.csv')
uscities
uscities_diversity = pd.read_csv('diversityWithCities4.csv')
uscities_diversity.drop(['Location', 'state'], axis=1, inplace=True)
uscities_diversity
```

```python
Pricedf = pd.read_csv("price.csv")
Pricedf
```

```python
Pricedf.dropna(how='all')
cityMatrix = Pricedf['City'].to_numpy()
valmatrix = Pricedf.iloc[:,6:]
valmatrix  = valmatrix.fillna(0)
valmatrix = valmatrix.to_numpy()
Price_means = []
for i in valmatrix:
    if i[0:].mean() == np.nan:
        print("ERROR")
    Price_means.append(i[0:].mean())

print(Price_means[0])   


tempcol = Pricedf.columns[6:]
Pricedf_with_means = Pricedf.drop(tempcol, axis=1)
Pricedf_with_means['PriceMean'] = Price_means
Pricedf_with_means
```

```python
masterdf = pd.merge(uscities_diversity, Pricedf_with_means, left_on=['city_ascii','county_name', 'state_id'], right_on=['City','County', 'State'], how='inner')
masterdf.drop(['City', 'Metro', 'County', 'City Code', 'State', 'Population Rank'], axis=1, inplace=True)
masterdf.columns
masterdf.to_csv('MasterDataset.csv', header=True, index=False)
```
